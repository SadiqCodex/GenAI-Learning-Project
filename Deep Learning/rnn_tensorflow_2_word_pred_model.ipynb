```python
quotes = df['quote']
quotes.head()
```

```python
quotes = quotes.str.lower()
```

```python
import string
translator = str.maketrans('','',string.punctuation)
quotes = quotes.apply(lambda x:x.translate(translator))
```

```python
quotes.head()
```

```python
vocab_size = 8978

tokinizer = Tokenizer(num_words=vocab_size)
tokinizer.fit_on_texts(quotes)
```

```python
word_index = tokinizer.word_index
print(len(word_index))
list(word_index.items())[0:10]
```

```python
sequences = tokinizer.texts_to_sequences(quotes)
for i in range(3):
    print(quotes[i])
```

```python
for i in range(3):
    print(sequences[i])
```

```python
X = []
Y = []
 
for seq in sequences:
    for i in range(1, len(seq)):
        input_seq = seq[:i]
        output_seq = seq[i]
        X.append(seq[:i])
        Y.append(seq[i])
```

```python
X
```

```python
Y
```

```python
print("X:",len(X),"Y:",len(Y))
```

```python
max_len = max(len(x) for x in X)
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_padded = pad_sequences(X, maxlen=max_len, padding='pre')
```

```python
X_padded
```

```python
y = np.array(Y)
```

```python
X_padded.shape
```

```python
from tensorflow.keras.utils import to_categorical

y_one_hot = to_categorical(y, num_classes=vocab_size)
```

```python
print(y_one_hot.shape)
print(y_one_hot[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM, SimpleRNN

embedding_dim = 50
rnn_units= 128

rnn_model = Sequential()
rnn_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim,input_length=max_len))

rnn_model.add(SimpleRNN(units=rnn_units))
rnn_model.add(Dense(units=vocab_size,activation='softmax'))
```

```python
rnn_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
rnn_model.summary()
```

```python
lstm_model = Sequential()
lstm_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim,input_length=max_len))

lstm_model.add(LSTM(units=rnn_units))
lstm_model.add(Dense(units=vocab_size,activation='softmax'))
lstm_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
lstm_model.summary()
```

```python
epochs=40
batch_size=128
```

```python
history_lstm=lstm_model.fit(
    X_padded,
    y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.1
)
```

```python
# Model Save
lstm_model.save("quote_lstm_model.h5")

print("Model saved successfully!")
```

```python
index_to_word = {}
for word, index in word_index.items():
    index_to_word[index] = word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model,tokenizer,text,max_len):
    text = text.lower()
    seq = tokenizer.texts_to_sequences([text])[0]
    seq = pad_sequences([seq], maxlen=max_len, padding='pre')
    pred = model.predict(seq,verbose=0)
    pred_index = np.argmax(pred)
    return index_to_word[pred_index]
```

```python
seed_text = "i love"
next_words = predictor(lstm_model,tokinizer,seed_text,max_len)
  
print(next_words)
```

```python
def generate_text(model,tokenizer,max_len,seed_text,n_words):
    for _ in range(n_words):
        next_word = predictor(model,tokenizer,seed_text,max_len)
        if next_word == "":
            break
        seed_text += " "+next_word
    return seed_text
```

```python
seed = "the meaning of life  "
generated_text = generate_text(lstm_model,tokinizer,max_len,seed,10)
print(generated_text)
```

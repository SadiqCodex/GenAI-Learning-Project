```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [10, 7, 20, 15, 30]

plt.figure(figsize=(8, 5))

plt.plot(
    days,
    sales,
    color='green',
    linestyle='dashed',
    linewidth=2,
    marker='o',
    markersize=8,
    label='Sales Data'
)

plt.title('Bakery Sales This Week')
plt.xlabel('Day of the Week')
plt.ylabel('Sales Per Day')
plt.legend(loc='upper right', bbox_to_anchor=(0, 1))
plt.grid(
    True,                 # grid on
    which='both',         # major + minor grid
    axis='both',          # x + y axis
    linestyle='--',       # dashed lines
    linewidth=0.7,        # line thickness
    alpha=0.6,           # transparency

color='blue')
plt.xlim(-0.5, 4.5)
plt.ylim(0, 35)
plt.xticks(days)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Group 1 (Low–Medium Hours)
hours_grp1 = [
1,1.2,1.4,1.6,1.8,
2,2.2,2.4,2.6,2.8,
3,3.2,3.4,3.6,3.8,
4
]

productivity_grp1 = [
4,5,6,6,7,
6,7,8,8,9,
7,8,9,9,10,
6
]

# Group 2 (High Hours)
hours_grp2 = [
4.2,4.4,4.6,4.8,
5,5.2,5.4,5.6,5.8,
6,6.2,6.4,6.6,6.8,
7,7.2,7.4,7.6,7.8,
8
]

productivity_grp2 = [
7,8,8,9,
7,8,9,9,10,
6,7,8,8,9,
7,8,9,9,10,
8
]

# Scatter Plot
plt.scatter(hours_grp1, productivity_grp1, marker='o', label='Group 1 (Low-Medium Hours)')
plt.scatter(hours_grp2, productivity_grp2, marker='^', label='Group 2 (High Hours)')

plt.xlabel('Hours Studied')
plt.ylabel('Productivity Level')
plt.title('Study Hours vs Productivity (Group Comparison)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Load CSV Data
# ---------------------------
df = pd.read_csv("netflix_dummy_data.csv")

# ---------------------------
# Remove missing values
# ---------------------------
df = df.dropna()

# ---------------------------
# Count Subscription Types
# ---------------------------
type_count = df["Subscription_Type"].value_counts()

# ---------------------------
# Plot Bar Chart
# ---------------------------
plt.figure(figsize=(8, 5))
plt.bar(type_count.index, type_count.values)

plt.title("Netflix Subscription Type Count (1 Year Data)")
plt.xlabel("Subscription Type")
plt.ylabel("Number of Users")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Load CSV File
# ---------------------------
df = pd.read_csv("netflix_dummy_data.csv")

# ---------------------------
# Remove Missing Values
# ---------------------------
df = df.dropna()

# ---------------------------
# Rating Count
# ---------------------------
rating_count = df["Rating_Given"].value_counts().sort_index()

# ---------------------------
# Pie Chart
# ---------------------------
plt.figure(figsize=(7, 7))

plt.pie(
    rating_count.values,
    labels=rating_count.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Netflix User Rating Distribution (1 Year Data)")
plt.xlabel("Rating Given")
plt.ylabel("Percentage of Users")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Load CSV
# ---------------------------
df = pd.read_csv("netflix_dummy_data.csv")

# ---------------------------
# Create Copy of DataFrame
# ---------------------------
data = df.copy()

# ---------------------------
# Clean Country Column
# ---------------------------
data["Country"] = data["Country"].str.strip()
data["Country"] = data["Country"].str.replace("USA", "United States")

# ---------------------------
# Convert Watch_Hours to Integer
# ---------------------------
data["Watch_Hours"] = data["Watch_Hours"].astype(int)

# ---------------------------
# Find Minimum Watch Hours
# ---------------------------
min_watch = data["Watch_Hours"].min()
print("Minimum Watch Hours:", min_watch)

# ---------------------------
# Histogram Plot
# ---------------------------
plt.figure(figsize=(10, 6))

plt.hist(
    data["Watch_Hours"],
    bins=10,
    edgecolor="black"
)

plt.title("Distribution of Watch Hours (Netflix One Year Data)")
plt.xlabel("Watch Hours per Day")
plt.ylabel("Number of Users")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Load CSV Data
# ---------------------------
df = pd.read_csv("netflix_dummy_data.csv")

# ---------------------------
# Select Required Columns
# ---------------------------
data = df[["Device_Type", "Subscription_Type"]].dropna()

# ---------------------------
# Convert Categorical to Numeric (Using Pandas)
# ---------------------------
device_map = {
    "Mobile": 1,
    "Tablet": 2,
    "Laptop": 3,
    "TV": 4
}

subscription_map = {
    "Basic": 1,
    "Standard": 2,
    "Premium": 3
}

data["Device_Code"] = data["Device_Type"].map(device_map)
data["Subscription_Code"] = data["Subscription_Type"].map(subscription_map)

# ---------------------------
# Scatter Plot
# ---------------------------
plt.figure(figsize=(8, 6))

plt.scatter(
    data["Device_Code"],
    data["Subscription_Code"]
)

plt.title("Device Type vs Subscription Type (Netflix Data)")
plt.xlabel("Device Type")
plt.ylabel("Subscription Type")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --------------------------------
# Load CSV Data
# --------------------------------
df = pd.read_csv("netflix_dummy_data.csv")

# --------------------------------
# Select Required Columns & Clean
# --------------------------------
data = df[["Country", "Subscription_Type"]].dropna().copy()

# --------------------------------
# GROUP 1 → Country Count
# --------------------------------
grp1 = data["Country"].value_counts().sort_values()

# --------------------------------
# GROUP 2 → Subscription Type Count
# --------------------------------
grp2 = data["Subscription_Type"].value_counts().sort_values()

# --------------------------------
# Create Subplots
# --------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --------------------------------
# Horizontal Bar Chart – Group 1
# --------------------------------
grp1.plot(
    kind="barh",
    ax=axes[0]
)

axes[0].set_title("Users by Country")
axes[0].set_xlabel("Number of Users")
axes[0].set_ylabel("Country")

# --------------------------------
# Horizontal Bar Chart – Group 2
# --------------------------------
grp2.plot(
    kind="barh",
    ax=axes[1]
)

axes[1].set_title("Users by Subscription Type")
axes[1].set_xlabel("Number of Users")
axes[1].set_ylabel("Subscription Type")

# --------------------------------
# Final Layout
# --------------------------------
plt.tight_layout()
plt.show()
```
